Column Descriptions
Job Title: The title of the job being advertised
Company: Name of the company posting the job
City: City where the job is located
Sector: Industry sector of the job
Salary Range PKR: Monthly salary range offered in Pakistani Rupees
Required Skills: Skills needed for the job
Experience Required: Years of experience required
Job Type: Whether job is Onsite, Hybrid or Remote
Education Level: Minimum education qualification required
Gender Preference: Whether employer prefers Male, Female or No Preference
Number of Vacancies: How many positions are available
Posted Date: Date when the job was posted
Application Deadline: Last date to apply for the job

```python
#IQR method
Q1 = df["Number of Vacancies"].quantile(0.25)
Q3  = df["Number of Vacancies"].quantile(0.75)
IQR = Q3-Q1

lower = Q1-1.5 * IQR
upper = Q3 +1.5 *IQR

outliers = df[df["Number of Vacancies"] > upper]
print(f"Total Outliers : {len(outliers)}")
print(f"Upper bound : {upper}")
```

8 specific questions about the dataset purely through code and print statements

```python
#Q1."Which city has the most job postings?"
print(df["City"].value_counts().idxmax())
```

```python
#Q2 "Which sector has the most job postings?"
print(df["Sector"].value_counts().idxmax())
```

```python
#Q3 "What percentage of jobs are remote?"

remote = df[df["Job Type"]=='Remote']
print(len(remote)/len(df) * 100)
```

```python
#Q4.  "Which sector has the highest average number of vacancies?"
print(df.groupby('Sector')['Number of Vacancies'].mean().idxmax())
```

```python
#Q5 "What is the most required education level?"
print(df['Education Level'].value_counts().idxmax())
```

```python
#Q6 "Which sector pays the highest salary?"

df['Salary Avg'] = df['Salary Range (PKR)'].str.replace(',', '').str.split(' - ').apply(lambda x: (int(x[0]) + int(x[1])) / 2 if isinstance(x, list) and len(x) == 2 else None)
print(df.groupby('Sector')['Salary Avg'].mean().idxmax())
```

```python
#Q7"What percentage of jobs prefer male candidates?"
male = df[df["Gender Preference"]=="Male"]
print(len(male)/len(df) * 100)
```

```python
#Q8 "Which city has the highest average salary?"
print(df.groupby('City')['Salary Avg'].mean().idxmax())
```

Data Quality Report
This dataset contains 10500 rows and 13 columns covering Pakistan job market data from May 2024 to May 2025 collected from major Pakistani hiring platforms.
Missing Values
Salary Range column has 3 percent missing values. Education Level has 2 percent missing values. Gender Preference has 1.5 percent missing values. Experience Required has 1 percent missing values. All missing values are under 5 percent so no column needs to be dropped. In Week 2 these will be filled using mode imputation.
Data Entry Errors
City column has serious spelling inconsistencies. Karachi is written as Karachì, karachi and KARACHI. Lahore appears as lahore, LAHORE and Lahorë. Islamabad appears as IslamaBad and islamabad. These are all the same cities but pandas treats them as different values which causes wrong counts. In Week 2 all city names will be standardized to proper title case.
Outliers
Number of Vacancies is the only numeric column. Using the IQR method the upper bound was calculated as 6. Any job posting with more than 6 vacancies is flagged as an outlier. Total 928 outlier rows were found. These are not necessarily errors since a company can genuinely need many hires at once. They will be investigated further in Week 2.
Redundant and Problematic Columns
Application Deadline column is redundant because Posted Date already provides time information. It will be dropped in Week 2. Required Skills column contains comma separated skill lists in a single cell which makes direct analysis impossible. In Week 2 this column will be split and exploded into individual rows for proper skill frequency analysis.
